```python
import matplotlib.pyplot as plt
```

```python
# Data to plot
labels = 'demolition permit granted', 'construction permit granted', 'construction started','construction not started'
sizes = [4047, 374, 3069, 116]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
#explode = (0, 0, 0, 0)  # explode 1st slice

fig1, ax1 = plt.subplots()

# Plot
ax1.pie(sizes, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')
plt.legend(labels)

plt.tight_layout()
plt.show()
```

```python
# Data to plot
labels = 'demolition permit granted = 4047', 'construction permit granted = 374'
sizes = [4047, 374]
colors = ['gold', 'lightskyblue']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors= colors, autopct='%1.1f%%', startangle=140)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax1.axis('equal')
plt.legend(labels, loc='upper right')
plt.tight_layout()
plt.title('Overview of construction and demolition status in netherlands');
plt.show()
```
